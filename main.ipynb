```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\project\swiggy.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
Abohar_data =df[df['city']== 'Abohar']
large5 = Abohar_data.nlargest(5, "rating")

fig = plt.figure(figsize = (25, 10))
 
# creating the bar plot
plt.bar(large5['name'], large5['rating'], color ='seagreen',
        width = 0.4)
 
plt.xlabel("Restaurants")
plt.ylabel("Rating")
plt.title("Abohar Top 5 rated Restaurants")
plt.show()
  
# display
# large5
```

```python
pd.to_numeric(df['id'], errors='coerce')
```

```python
df['city'] = df['city'].apply(lambda x: x.split(',')[-1])

fig = plt.figure(figsize = (25, 10))          
sns.countplot(x="city",data=df, palette="Greens_d",
              order=df.city.value_counts().iloc[:5].index)
plt.xlabel("City")
plt.title("Most order citys ")
plt.show()
  
```

```python
most_order_rest = df.groupby(['name'])['id'].count().sort_values(ascending=False).head(5)

fig = plt.figure(figsize = (25, 10)) 
# creating the bar plot
most_order_rest.plot(kind = 'bar')
 
plt.xlabel("Restaurants")
plt.ylabel("Rating")
plt.title("Top 5 Restaurants order")
plt.show()

```
